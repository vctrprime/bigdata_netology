```python
import pandas as pd
```

```python
data = pd.read_csv('abalone.csv')
```

```python
#круговая диаграмма распеделения по полу
import matplotlib.pyplot as plt

males = (data["Sex"] == 'M').sum()
females = (data["Sex"] == 'F').sum()
infants = (data["Sex"] == 'I').sum()

proportions = [males, females, infants]

plt.pie(
    proportions,
    labels = ['M', 'F', 'I'],
    shadow = False,
    colors = ['blue', 'red', 'green'],
    startangle = 90,
    explode = (0.05, 0.05, 0.05),
    autopct = '%1.1f%%'
)

plt.axis('equal')
plt.title('Распределение по полу')
plt.show()
```

```python
#скаттер диаметр-пол-кольца (первые сто)
import seaborn as sns
abalones = data[:100]
lm = sns.lmplot(x = 'Rings', y = 'Diameter', data = abalones, hue = 'Sex', fit_reg=False)
lm.set(title='Диаметр-Пол-Кольца')
```
